```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definindo as equaes do sistema
def eq1(x, z):
    return 1 - x

def eq2(x, y):
    return 3 - x - y

# Criando pontos para plotar os planos das equaes
x = np.linspace(-10, 10, 100)
z = np.linspace(-10, 10, 100)
X, Z = np.meshgrid(x, z)
Y1 = eq1(X, Z)
Y2 = eq2(X, Z)

# Plotando os planos das equaes
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y1, Z, alpha=0.5, rstride=100, cstride=100)
ax.plot_surface(X, Y2, Z, alpha=0.5, rstride=100, cstride=100)

# Adicionando a linha que representa y = 2
y = np.ones_like(x) * 2
ax.plot(x, y, zs=0, zdir='y', color='red')

# Configuraes adicionais do grfico
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title('Grfico 3D das Equaes do Sistema')
plt.show()
```
